```python
"""
Question 1

A game is played whereby a player: 
- flips one fair coin
- rolls one regular fair (six-sided) die 
- and selects one card uniformly at random from a standard deck of 52 cards.
"""

def coin_flip():
    """
    one point each time the coin lands Heads and
    zero each time it lands Tails
    RETURN score
    """
    if random.random() <= 0.5:
#         Heads
        return 1
    else:
#         Tails
        return 0

def dice_throw():
    """
    one point for each number shown on the die 
    (i.e. if the die shows four, he/she scores four points)
    RETURN score
    """
    x = random.random()
    for i in range(1,7):
        if x <= i/6:
            return i

def card_draw():
    """
    one point each time the card drawn is Clubs or Spades, 
    four points each time the card drawn is Hearts and 
    zero each time the card drawn is Diamonds.
    RETURN score
    """
    x = random.random()
    if x <= 0.5:
#         Black Card
        return 1
    elif x <= 0.75:
#         Heart
        return 4
    else:
#         Diamond
        return 0

def play_game():
    return coin_flip() + dice_throw() + card_draw()

# Number of simulations
n = 2000
# The probability that the player scores exactly one point in total
i = 0
# The probability that the player scores an even number of points in total
ii = 0
# The probability that the player obtains the same score when playing the game twice in a row
iii = 0
for _ in range(n):
    score = play_game()
    
    if score == 1:
        i += 1
        
    if score %2 == 0:
        ii += 1
        
    if score == play_game():
        iii += 1
    
print(f"The probability that the player scores exactly one point in total: {i/n}")
print(f"The probability that the player scores an even number of points in total: {ii/n}")
print(f"The probability that the player obtains the same score when playing the game twice in a row: {iii/n}")
```

```python
"""
Question 2

Status of library book is (borrowed or available) is monitored each day.

IF Borrowed:
    - Borrowed tomorrow = 80%
    - Available tomorrow = 20%

IF Available:
    - Borrowed tomorrow = 35%
    - Available tomorrow = 65%

Simulate 2000 days of the the books status.

Estimate the probabiliity that the book is available at a given day in the study period.
"""

# Number of simulations
n = 2000
days = [True] # True is avaliable
for i in range(n-1):
    x = random.random()
#     If it is borrowed when checked, 
    if days[i] == False:
#          then the probability that it is still borrowed the following day is 80%
        if x <= 0.8:
            days.append(False)
#          and the probability that it has been returned is 20%.
        else:
            days.append(True)

#     If it is available when checked, 
    else:
#          then it is borrowed a day later with probability 35% 
        if x <= 0.35:
            days.append(False)
#          and remains available a day later with probability 65%.
        else:
            days.append(True)


print(f"The estimated probability the book is available at a given day in the study period: {(days.count(True)/n)}")
```

```python
"""
Question 3

At each game, each of the two players select one of three options: 
rock, paper or scissors.

If both players select the same option, the game is drawn.

If they are different, then:
    - a player selecting rock wins against a player selecting scissors. 
    - a player selecting scissors wins against a player selecting paper. 
    - a player selecting paper wins against a player selecting rock.

Both players select rock for the first two games.

For subsequent games, player A’s selections are random.
    - He/she selects rock for each game with probability 0.6 
    - and selects either of the other two options with probability 0.2. 

Player B simply decides to play whichever move would beat player A 
if he/she chose the same option as he/she did two games ago. 
For example, if in game 10, player A selects rock, 
then player B will select paper in game 12.

Simulate this scenario 10,000 times and hence estimate what proportion of games 
will be drawn and what proportion will be won by both player A and by player B.

Comment on the result.
"""

show_logs = False

def r_p_s(A,B):
    # If both players select the same option, the game is drawn.
    if A == B:
        return 'draw'
    
    # If they are different, then:
    #     - a player selecting rock wins against a player selecting scissors. 
    #     - a player selecting scissors wins against a player selecting paper. 
    #     - a player selecting paper wins against a player selecting rock.
    if (A == 'sissors' and B == 'paper') or (A == 'paper' and B == 'rock') or (A == 'rock' and B == 'sissors'):
        return 'A'
    else:
        return 'B'

# Player B simply decides to play whichever move would beat player A 
# if he/she chose the same option as he/she did two games ago. 
# For example, if in game 10, player A selects rock, 
# then player B will select paper in game 12.
def player_b_choice(A):
    if A == 'sissors':
        return 'rock'
    elif A == 'paper':
        return 'sissors'
    else:
        return 'paper'

    
# Number of simulations
N = 10000

# Both players select rock for the first two games.
player_A = ['rock','rock']
player_B = ['rock','rock']

if show_logs:
    print(f"Player A\n{player_A}\nPlayer B\n{player_B}\n")

for n in range(2,N):
    if show_logs:
        print(n)
        
    x = random.random()
    
#     For subsequent games, player A’s selections are random.
#     - He/she selects rock for each game with probability 0.6 
#     - and selects either of the other two options with probability 0.2. 
    if x <= 0.6:
        A = 'rock'
    elif x <= 0.8:
        A = 'paper'
    else:
        A = 'sissors'
    
    B = player_b_choice(player_A[n-2])
    
    if show_logs:
        print(f"Player A 2 games ago: {player_A[n-2]} | n-2 = {n-2} | B = {B}")
        print(f"A: {A} | B: {B}")
        
    player_B.append(B)
    player_A.append(A)
    
games = []
for i in range(N):
    games.append(r_p_s(player_A[i], player_B[i]))    

if show_logs:
    print(f"Player A\n{player_A}\nPlayer B\n{player_B}\nGames\n{games}")
    print(f"Sense check: {(games.count('draw')/N) + (games.count('A')/N) + (games.count('B')/N)} == 1")

print(f"The estimated proportion of games:\n\tDrawn: {(games.count('draw')/N)}\n\tWon by Player A: {(games.count('A')/N)}\n\tWon by Player B: {games.count('B')/N}")
print("Because A chooses rock 60% of the time his/her choice is slightly more predictable giving B an advantage.")
```
