```python
"""
1.
A gambler bets repeatedly on a standard (European-style) roulette wheel. 

Each time he/she bets $1 and:
- gains an additional $1 with probability 18/37
- and loses his/her stake with probability 19/37 . 

The gambler begins with $15 
and bets repeatedly until he/she:
- first reaches $20 (wins) 
- or reaches $0 (loses.)

Simulate 100 full runs of the gambler’s games until he/she reaches 
either absorbing boundary. 

From these simulations, estimate:
i) The probability that the gambler reaches $20 and not $0;
ii) The probability that the gambler places more than 50 bets before
reaching either boundary;
iii) The expected number of bets placed before reaching either boundary;
iv) The expected number of times that the gambler has exactly $17 during each full game.

Note: 100 full runs of the game requires simulation of 100 times that 
the player’s money begins at $15 and ends at either $0 or $20, not just 100 bets.
"""

def roulette():
    x = random.uniform(0,1)
# Each time he/she bets $1 and:
# - gains an additional $1 with probability 18/37
# - and loses his/her stake with probability 19/37 . 
    if x <= (18/37):
        return 1
    else:
        return -1

show_logs = False
N = 100
win = 20
lose = 0
start = 15
num_games = []
results = [] # True == win
expect_k = 17
expect_k_s = []

for _ in range(N):
    k = start
    game_count = 0
    found_expect_k = 0

    while k < win and k > lose:
        k += roulette()
        game_count += 1
        if k == expect_k:
            found_expect_k += 1

    if show_logs:
        print(f"k: {k}")

    results.append(k == win)
    num_games.append(game_count)
    expect_k_s.append(found_expect_k)

print(f"i) The probability that the gambler reaches $20 and not $0: {results.count(True)/N}")

games_more_50 = 0
for i in num_games:
    if i > 50:
        games_more_50 += 1
        
print(f"ii) The probability that the gambler places more than 50 bets before reaching either boundary: {games_more_50/N}")

print(f"iii) The expected number of bets placed before reaching either boundary: {mean(num_games)}")

print(f"iv) The expected number of times that the gambler has exactly $17 during each full game: {mean(expect_k_s)}")
```

```python
"""
2. A game is played whereby a player rolls a regular fair six-sided die. 

- If the die shows an even number, the player’s score is zero. 
- If the die shows an odd number, the player scores the number of points shown, 
    i.e. if the die shows a 5, the player scores 5 points.
    
Simulate 1000 realisations of this game.

i) From your simulations, estimate the expected number of points scored in a single game.

The player is now given a call option, allowing (but not obliging) 
the player to swap his/her prize for a prize of 2 points. 
This call can be made after the die roll is observed and the player will 
always act to maximise his/her score.

ii) From your simulations, estimate the fair price for this call option.
"""

def dice_throw():
    x = random.random()
    for i in range(1,7):
        if x <= i/6:
            return i

def get_score_i(x):
    if x %2 == 0:
        return 0
    else:
        return x

def get_score_ii(x):
    if x < 2:
        return 2
    else:
        return x
    
N = 1000
results_i = []
results_ii = []

for _ in range(N):
#     print(f"{get_score(dice_throw())}")
    score = get_score_i(dice_throw())
    results_i.append(score)
    results_ii.append(get_score_ii(score))

print(f"i) Expected number of points scored in a single game: {mean(results_i)}")
print(f"ii) From your simulations, estimate the fair price for this call option: {round(mean(results_ii) - mean(results_i),3)}")
```

```python
"""
3. Consider an undirected random graph on the set of four nodes {A, B, C, D} 
such that each of the 4C2 = 6 potential edges exists with probability 0.3, 
independently of the presence/absence of any other edges.

A graph is connected if any node can be reached along an edge or 
combination of edges starting from any other. 

For example, if edges AC, AD and BD were present, 
but potential edges AB, BC and CD were absent, 
the graph would still be connected, 
since any node can still be reached from any other.

Simulate 1000 such random graphs.
i) From your simulations, estimate the probability that such a random graph is connected

The shortest path length between two nodes is defined as smallest 
number of edges which can be used to link the two nodes. 

For example, if edge AB exists, then the shortest path length between A and B is 1. 

If AB does not exist, but edges AC and BC exist, then the shortest path length is 2. 

The shortest path length between two nodes on a disconnected graph may be infinite.

ii) From your simulations, estimate the expected shortest path length between two nodes, 
excluding cases where this value is infinite i.e. excluding disconnected graphs.
"""

show_logs = False
N = 1000
num_connected = 0
edge_probability = 0.3
graph_size = 4
pool = ['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O']
path_lengths = []

for _ in range(N):
    # Build Graph
    graph = pool[:graph_size]
    edges = [ [ 0 for i in range(graph_size) ] for j in range(graph_size) ]

    for i in range(graph_size):
        for j in range(i,graph_size):
            if i == j:
                edges[i][j] = 1
                edges[j][i] = 1
                
            if np.random.uniform(0,1) <= edge_probability:
                edges[i][j] = 1
                edges[j][i] = 1

            if show_logs:
                print(f"col: {i} row: {j} | {graph[i]} -> {graph[j]} : {edges[i][j]} == {edges[j][i]}")

    # Check Connected
    tmp  = pool[:graph_size]
    for i in range(len(graph)):
        
        if show_logs:
            print(edges[i])
        
        for j in range(len(graph)):
            if i != j and edges[i][j] == 1 and graph[j] in tmp:
                tmp.remove(graph[j])

    if show_logs:
        print(f'{tmp}')
    
    if len(tmp) == 0:
        num_connected += 1
        
        for i in range(len(graph)):
            path_length = 0
            if show_logs:
                print(f"from: {graph[i]}")
            for j in range(i,len(graph)):
                path_length += 1
                if show_logs:
                    print(f"to: {graph[j]} = {edges[i][j]}")
                if i != j and edges[i][j] == 1:
                    path_lengths.append(path_length)
                    if show_logs:
                        print(f"{path_length}")
        
        
print(f"i) Estimated probability that such a random graph is connected: {round(num_connected/N*100,2)}%")
print(f"ii) From your simulations, estimate the expected shortest path length between two nodes, excluding cases where this value is infinite i.e. excluding disconnected graphs: {round(mean(path_lengths),2)}")
```
