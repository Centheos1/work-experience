```python
import numpy as np
import math
from statistics import *
```

```python
def card_draw():
    """
    one point each time the card drawn is Clubs or Spades, 
    four points each time the card drawn is Hearts and 
    zero each time the card drawn is Diamonds.
    RETURN score
    """
    x = np.random.uniform(0,1)
    if x <= 0.25:
        return 'heart'
    elif x <= 0.5:
        return 'diamond'
    elif x <= 0.75:
        return 'club'
    else:
        return 'spades'
    
```

```python
# Question 1

logs = False
N = 2000 # number of simulations
n = 20 # number of draws
num_streaks = []

for _ in range(N):
    draws = []
    for _ in range(n):
        draws.append(card_draw())


    consecutive = 0
    tmp = 1

    for i in range(1, len(draws)):
        if draws[i - 1] == draws[i]:
            tmp += 1
        else:
            if tmp > consecutive:
                consecutive = tmp
            tmp = 1
            
    if logs:
        print(f"consecutive: {consecutive}")
    
    num_streaks.append(consecutive)
    
for i in range(1,n+1):
    print(f"The estimated proportial having a longest streak of {i} is {num_streaks.count(i) / N}")
    
```

```python
# Question 2
logs = True

def run_test(observations):
    hypothesis = False
    significance = 1.96 # N(0,1)
    E = ( 2 * len(observations) - 1 ) / 3
    Var = ( 16 * len(observations) - 29) / 90
    A = 0
    curr = None # Positive == True
    prev = None
    
    for i in range(1, len(observations)):
        
        if prev is None:
            prev = observations[i - 1] > observations[i]
        
        curr = observations[i - 1] > observations[i]
        
        if curr != prev:
            A += 1
        
        prev = curr
        
    Z = (A - E) / math.sqrt(Var)

    if abs(Z) < significance:
        hypothesis = True
        
    print(f"Run Test:\n\tE = {E}\n\tVar = {Var}\n\truns = {A}\n\tZ = {Z}\n\tIs random = {hypothesis}")

def run_test_relative_to_mean(observations):
    hypothesis = False
    significance = 1.96 # N(0,1)
    B = 0
    curr = None # Positive == True
    prev = None
    
    N_pos = 0
    N_neg = 0
    
    for i in range(len(observations)):
        
        if observations[i] > mean(observations):
            N_pos += 1
        
        if observations[i] < mean(observations):
            N_neg += 1
        
        if i != 0:
            if prev is None:
                prev = observations[i - 1] > mean(observations)

            curr = observations[i - 1] > mean(observations)

            if curr != prev:
                B += 1

            prev = curr
    
    E = 1/2 + ( 2 * N_pos * N_neg /  len(observations) )
    
    Var = (2 * N_pos * N_neg * (2 * N_pos * N_neg - len(observations)) ) / (len(observations)**2 * (len(observations) - 1))
    
    Z = (B - E) / math.sqrt(Var)
    
    if abs(Z) < significance:
        hypothesis = True
    
    print(f"Run Test Relative to Mean:\n\tE = {E}\n\tVar = {Var}\n\truns = {B}\n\tZ = {Z}\n\tIs random = {hypothesis}")

    
def linear_congruence_generator(x_0, a, c, m, n):
    x_vals = []
    if logs:
        print(f"Linear Congruence Generator:\n\tseed (x_o): {x_0}\n\tmultiplier (a): {a}\n\tincrement (c): {c}\n\tmodulus (m): {m}")
    for _ in range(n):
        x_n = (a * x_0 + c) % m
        x_vals.append(x_n)
        x_0 = x_n
    return x_vals
    
```

```python
a = 65539 # multiplier
c = 0 # increment
m = 2**31 # modulus
seed = 11011 # start value
n = 1000 # number of values

observations = linear_congruence_generator(seed, a, c, m, n)
random_numbers = []
for i in observations:
    random_numbers.append(i / m)

run_test(random_numbers)
run_test_relative_to_mean(random_numbers)
```
