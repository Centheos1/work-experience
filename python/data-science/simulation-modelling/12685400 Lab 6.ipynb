```python
# iii
realisations = []
N = 20

def simulation():
    
    mu_s = [False, 4, 8, 12]
    lambda_s = 2
#     mu_s = [False, 1, 2, 3]
    lambda_s = 1
    state_history = []
    
    for t in range(timesteps):
        state = 0
        
        state_history.append([])

        if np.random.uniform(0,1) < lambda_s * time_step:
            if(state < 3):
                state+=1
        if(mu_s[state]):
            if(np.random.uniform(0,1)<mu_s[state]*time_step):
                state -=1 
    #     state_history[t].append(state)
        state_history[t] = state
        
    
    return state_history

for i in range(N):
    realisations.append(simulation())
    
timeline = []
for i in range(timesteps):
    timeline.append(i)

state_0 = [0] * len(realisations[0])
state_1 = [0] * len(realisations[0])
state_2 = [0] * len(realisations[0])
state_3 = [0] * len(realisations[0])

data = {
    "timeline" : timeline,
    "state_0" : state_0,
    "state_1" : state_1,
    "state_2" : state_2,
    "state_3" : state_3
}

for j in realisations:
    time_in_0 = 0
    time_in_1 = 0
    time_in_2 = 0
    time_in_3 = 0
    
# #     print(i)
    for i in range(len(j)):
    
        if j[i] == 0:
            time_in_0 += 1

        if j[i] == 1:
            time_in_1 += 1

        if j[i] == 2:
            time_in_2 += 1

        if j[i] == 3:
            time_in_3 += 1
            
        data['state_0'][i] += time_in_0/(i + 1)
#         print(time_in_0/(i + 1))
        data['state_1'][i] += time_in_1/(i + 1)
        data['state_2'][i] += time_in_2/(i + 1)
        data['state_3'][i] += time_in_3/(i + 1)
#         print(time_in_3/(i + 1))
        

for i in range(len(data['state_0'])):
    data['state_0'][i] = data['state_0'][i] / N
    data['state_1'][i] = data['state_1'][i] / N
    data['state_2'][i] = data['state_2'][i] / N
    data['state_3'][i] = data['state_3'][i] / N
    
print(f"time_in_0: {time_in_0}")
print(f"time_in_1: {time_in_1}")
print(f"time_in_2: {time_in_2}")
print(f"time_in_3: {time_in_3}")
    
```
