```python
import sympy as sp
from sympy import *
import numpy as np
import math
from statistics import *

import matplotlib.pyplot as plt   
from matplotlib.pyplot import *
```

```python

# Question 1

# a

# i)
logs = False
N = 20
realisations = []
end_time = 5
h = 0.002 # time step in minutes

timesteps = round(end_time/h)
# timesteps

time_view = int(10/60 * 1 / h) # every 10 seconds

for j in range(N):
    realisations.append([])
    n = 1 # population size
    t = 0 # time
    _lambda = 2 * n * h
    
    for i in range(timesteps):
    #     print(t)
    #     is_birth
        if np.random.uniform(0,1) < ( 2 * n * h ):
            n += 1
    #         print(n)

        if i % time_view == 0:
            if logs:
                print(f"time: {t * 60} | population: {n}")
            realisations[j].append(n)
#             timeline.append(i)

        t += h

# n
data_visualisation = []
len(realisations)
for i in range(len(realisations[0])):
    x = 0
    for j in range(N):
#         print(f"i:{i} j:{j} {realisations[j][i]}")
        x += realisations[j][i]
    
    print(f"Expected population at {(i+1)*10} seconds: {x/N}")
    data_visualisation.append(x/N)
    
    
# ii)
ii = """
ii) Because the probability of a birth is dependand on the population size and the time step length
once the population size and time step gets large enough the probability of a birth will be greater than 1 and
grows linearly not exponentially
"""
print(ii)
```

```python
timeline = []
for i in range(len(data_visualisation)):
    timeline.append(i)

figure(figsize=(16, 9), dpi=80)

x = timeline
y1 = data_visualisation

plt.plot(x,y1, color='b', label='Expected Population')

# Turn a grid on.
plt.grid()

# Title the plot.
plt.title('Time in State')


# Label axes
plt.xlabel('Time Step (Every 10 seconds)')
plt.ylabel('Population Size')

# Make the plot have a legend based on the data 'labels'
plt.legend()
```

```python
# b)

logs = False
N = 20
data_visualisation = []
realisations = []
K = 200
end_time = 5
h = 0.002 # time step in minutes


timesteps = round(end_time/h)
# timesteps

time_view = int(10/60 * 1 / h) # every 10 seconds

for j in range(N):
    realisations.append([])
    n = 1 # population size
    t = 0 # time
    _lambda = 2 * n * h
    
    for i in range(timesteps):
    #     print(t)
    #     is_birth
        if np.random.uniform(0,1) < ( 2 * ( 1 - (n/K) ) * n * h ):
            n += 1
    #         print(n)

        if i % time_view == 0:
            if logs:
                print(f"time: {t * 60} | population: {n}")
            realisations[j].append(n)

        t += h

# n
len(realisations)
for i in range(len(realisations[0])):
    x = 0
    for j in range(N):
#         print(f"i:{i} j:{j} {realisations[j][i]}")
        x += realisations[j][i]
    
    print(f"Expected population at {(i+1)*10} seconds: {x/N}")
    data_visualisation.append(x/N)
    
    
# ii)
ii = """
ii) Because the probability of a birth is approaches 0 as time approaches infinity the model 
is more realistically allowing for the limited resources of real life that prevent populations
size growing expoentally.
"""
print(ii)
```
